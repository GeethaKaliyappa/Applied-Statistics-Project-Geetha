#### 2A) Probability that none of the items are defective?

```python
from scipy.stats import binom

x  =  0.05 # failure rate is 5%
n   =  10 # sample size
k   =  np.arange(0,11) 

bino = binom.pmf(k,n,x)
print(bino)

print('Probability that none of the items are defective: %1.6f' %bino[0])
```

#### 2B) Probability that exactly one of the items is defective?

```python
print('Probability that exactly one of the items is defective: %1.6f' %bino[1])
```

#### 2C) Probability that two or fewer of the items are defective?

```python
cdf = binom.cdf(k,n,x)
print(cdf)
print('Probability that two or fewer of the items are defective: is %1.6f' %cdf[2])
```

#### 2D) Probability that three or more of the items are defective ?

```python
P = 1- cdf[2]
print('Probability that three or more of the items are defective: %1.6f' % P)
```

#### 3A) What is Probability that in a given week he will sell some cars?

```python
from scipy.stats import poisson

n=np.arange(0,15)

a=3

P=(poisson.pmf(n, a))

print(P)

P1=(1-P[0])

print("\nThe probability of the salesman selling some cars in a week is: ", '%1.6f' %P1)
```

#### 3B) What is Probability that in a given week he will sell 2 or more but less than 5 cars? [1 Mark]

```python
#Probability for 2 cars, probability for 3 cars, and probability for 4 cars

P2=(P[2]+P[3]+P[4])

print(" Probability that in a given week he will sell 2 or more but less than 5 cars: ", '%.6f' %P2)
```

**C. Test for significance and conclusion**

Alpha value = 0.05.

since p_value is greated than Alpha, we fail to reject the null hypothesis.


So, there is no significant difference between funds raised by companies that are still operating and the companies that closed down

**G. Write your hypothesis comparing the proportion of companies that are operating between winners and contestants:**

**#Null Hypothesis**

**#Alternate Hypothesis**

**H0** = There is no significant difference between the proportion of operating companies that are winner and Contestants (p1=p2)

**H1** = There is a significant difference between the proportion of operating companies that are winner and Contestants (p1≠p2)

Level of Significance = Alpha(α) = 0.05

**H. Test for significance and conclusion.**

```python
#testing of proportions

n1 = 662
n2 = 662

p1 = 0.272
p2 = 3.233

x1= 18
x2= 214

#Alpha_1 = 0.025 = +1.96
#Alpha_2 = 0.025 = -1.96
```

```python
import math

#testing of proportions

p_hat = (x1+x2)/(n1+n2)

print("p_hat= ",p_hat)

numerator= (p1-p2)-0

denominator= p_hat*(1-p_hat)*(2/662)

test_stat = numerator/math.sqrt(denominator)

print('\ntest_stat= ', '%.2f' % test_stat)
```


Since the test_stat value i.e. -141.70 is falling in the rejection region i.e over -1.96,
